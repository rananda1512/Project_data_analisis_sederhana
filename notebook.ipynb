## Exploratory Data Analysis (EDA)

### Explore ...

```python
# melihat cuaca mana yang banyaknya penyewa sepeda
df_whether=df_day.groupby(by="weathersit").cnt.sum().sort_values().reset_index()
df_whether
```

```python
# melihat musim mana yang banyaknya penyewa sepeda
df_season=df_day.groupby(by="season").cnt.sum().sort_values(ascending=False).reset_index()

# membuat nilai interger season menjadi string agar mudah dipahami audien
df_season['season'] = df_season['season'].apply(lambda x: 'Musim Dingin' if x == 1 
                                                     else 'Musim Semi' if x == 2 
                                                     else 'Musim Panas' if x == 3 
                                                     else 'Musim Gugur')
df_season
```

```python
# melihat perbandingan antara penyewa casual dengan register dalam 1 musim
df_season_casual_register=df_day.groupby(by="season").agg({
    "casual":"sum",
    "registered":"sum",
}).sort_values(by="casual").reset_index()

# membuat nilai interger season menjadi string agar mudah dipahami audien
df_season_casual_register['season'] = df_season_casual_register['season'].apply(lambda x: 'Musim Dingin' if x == 1 
                                                     else 'Musim Semi' if x == 2 
                                                     else 'Musim Panas' if x == 3 
                                                     else 'Musim Gugur')
df_season_casual_register
```

```python
# Melihat jam berapakah yang banyak menyewa sepeda
df_hour_count=df_hour.groupby(by="hr").cnt.sum().sort_values()

df_hour_count
```

**Insight:**
- Cuaca cerah merupakan cuaca terbaik untuk menyewa sepeda
- Musim panas menjadi musim terbanyak penyewa sepeda
- Penyewa register lebih banyak dari penyewa casual untuk setiap musim
- Jam 17 atau 5 sore merupakan pilihan terbaik untuk meyewa sepeda karena bisa dipakai untuk olahraga sekaligusxx

## Visualization & Explanatory Analysis

### Pertanyaan 1: Kapan terakhir jumlah penyewaan sepede mencapai angka tertinggi dalam sehari

```python
# Menghitung total penyewaan per hari
df_daily_counts = df_day.groupby('dteday')['cnt'].sum().reset_index()

# Mencari hari dengan jumlah penyewaan tertinggi
max_count_day = df_daily_counts.loc[df_daily_counts['cnt'].idxmax()]

print(f"Hari dengan jumlah penyewaan tertinggi adalah {max_count_day['dteday']} dengan jumlah {max_count_day['cnt']}.")

plt.figure(figsize=(12, 6))
plt.plot(df_daily_counts['dteday'], df_daily_counts['cnt'], marker='o', color='#87A2FF')
plt.xticks(rotation=45)
plt.title('Tren Penyewaan Sepeda per Hari')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Penyewaan')
plt.grid()
plt.show()
```

### Pertanyaan 2: Berapa banyak penyewaan yang dilakukan oleh setiap pengguna selama periode waktu tertentu?

```python
plt.figure(figsize=(10, 6))
plt.bar(df_season['season'], df_season['cnt'], color='skyblue')
plt.title('Jumlah Penyewaan Sepeda Berdasarkan Musim')
plt.xlabel('Musim')
plt.ylabel('Jumlah Penyewaan')
plt.xticks(rotation=45)  # Memutar label sumbu x agar lebih mudah dibaca
plt.show()
```

```python
season_input = int(input("Masukkan nomor musim (1: Musim Dingin, 2: Musim Semi, 3: Musim Panas, 4: Musim Gugur): "))

season_name = df_season_casual_register[df_season_casual_register['season'].str.contains(
    'Musim Dingin' if season_input == 1 
    else 'Musim Semi' if season_input == 2 
    else 'Musim Panas' if season_input == 3 
    else 'Musim Gugur'
)].iloc[0]

# Membuat Pie Chart untuk musim yang dipilih
labels = ['Pengguna Kasual', 'Pengguna Terdaftar']
sizes = [season_name['casual'], season_name['registered']]
colors = ['red', 'blue']
explode = (0.1, 0)  # Memisahkan bagian pengguna kasual sedikit dari pie

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is a circle.
plt.title(f'Perbandingan Penyewaan Sepeda di {season_name["season"]}')
plt.show()
```

**Insight:**
- Dari pertanyaan 1 kita bisa lihat terjadinya penurunan yang sangat dratis di sekitaran tahun 2013
- Pertanyaan 2 sangat cocok menggunakan bar diagram karena membandingkan lebih dari dua nilai
- Pertanyaan 3 sangat cocok menggunakan pie diagram karena hanya membandingkan perbedaan dua nilai

## Conclusion

- Kesimpulan untuk pertanyaan pertama adalah terjadinya penurunan penyewaan sepeda di sekitaran tahun 2013
